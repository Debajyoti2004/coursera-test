```python
def get_tokenized_data(file):
  def preprocessing_data(file):
    with open(file, 'r') as f:
      data = f.read()
    sentences=data.split('\n')
    sentences=[s.lower() for s in sentences]
    sentences=[s for s in sentences if len(s)>0]
    sentences=[re.sub('[^a-zA-Z0-9?.!,;]',' ',s) for s in sentences]
    return sentences

  def tokenized_sentences(sentences):
    tokenized_sentences=[]
    for sentence in sentences:
      tokenized_sentence=nltk.word_tokenize(sentence)
      tokenized_sentences.append(tokenized_sentence)
    return tokenized_sentences

  sentences=preprocessing_data(us_twitter_txt_path)
  tokenized_sentences=tokenized_sentences(sentences)
  return tokenized_sentences
```

```python
def get_word_with_freq(tokenized_sentences,count_threhold=2):
  def word_counts(tokenized_sentences):
    word_counts={}
    for sentence in tokenized_sentences:
      for word in sentence:
        if word not in word_counts:
          word_counts[word]=1
        else:
          word_counts[word]+=1
    return word_counts

  def filter_words(word_counts,count_threhold):
    filtered_words={k:v for k,v in word_counts.items() if v>=count_threhold}
    return filtered_words

  word_counts=word_counts(tokenized_sentences)
  filtered_words=filter_words(word_counts,count_threhold)
  closed_vocab=list(filtered_words.keys())
  return closed_vocab,filtered_words
```

```python
def replace_oov_words_with_unk(tokenized_sentences, closed_vocab):
  replaced_sentences=[]
  vocabulary=set(closed_vocab)

  for sentence in tokenized_sentences:
    replaced_sentence=[]
    for word in sentence:
      if word in vocabulary:
        replaced_sentence.append(word)
      else:
        replaced_sentence.append('<unk>')
    replaced_sentences.append(replaced_sentence)
  return replaced_sentences
```

```python
def split_data(tokenized_sentences, split_ratio=0.8):
  random.seed(42)
  random.shuffle(tokenized_sentences)
  split_index=int(len(tokenized_sentences)*split_ratio)
  train_sentences=tokenized_sentences[:split_index]
  test_sentences=tokenized_sentences[split_index:]
  return train_sentences,test_sentences
```

```python
from collections import defaultdict

def count_n_grams(data, n, start='<s>', end='<e>'):
  n_grams = defaultdict(int)
  for sentence in data:
    sentence = [start] * n + sentence + [end]
    for i in range(len(sentence) - n + 1):
      n_gram = tuple(sentence[i:i+n])
      n_grams[n_gram] += 1
  return n_grams

def estimate_probability(word, n_gram_counts, nplus1_gram_counts, previous_n_gram, n, vocabulary_size, k=1.0):
  nplus1_gram = previous_n_gram + (word,)
  numerator = nplus1_gram_counts.get(nplus1_gram, 0) + k
  denominator = n_gram_counts.get(previous_n_gram, 0) + k * vocabulary_size
  probability = numerator / denominator
  return probability
```

```python
def estimate_probabilities(previous_n_gram, n, n_gram_counts, nplus1_gram_counts, vocabulary, k=1.0):
  previous_n_gram = tuple(previous_n_gram)
  vocabulary = vocabulary + ["<e>", "<unk>"]
  vocabulary_size = len(vocabulary)
  probabilities = {}
  for word in vocabulary:
    probability = estimate_probability(word, n_gram_counts, nplus1_gram_counts, previous_n_gram, n, vocabulary_size, k=k)
    probabilities[word] = probability
  return probabilities
```

```python
def create_probablility_matrix(n_pus_1_grams_counts,k,vocabulary):
  def create_count_matrix(n_pus_1_grams_counts,vocabulary):
    vocabulary=vocabulary+["<e>","<unk>"]
    n_grams=[]
    for key in n_pus_1_grams_counts:
      n_grams.append(key[:-1])
    n_grams=list(set(n_grams))
    row_index={n_gram:i for i,n_gram in enumerate(n_grams)}
    col_index={word:j for j,word in enumerate(vocabulary)}
    nrow=len(n_grams)
    ncol=len(vocabulary)

    count_matrix=np.zeros((nrow,ncol))
    for key,value in n_pus_1_grams_counts.items():
      n_gram=key[:-1]
      word=key[-1]
      if word not in vocabulary:
        continue
      i=row_index[n_gram]
      j=col_index[word]
      count_matrix[i,j]=value
    count_matrix=pd.DataFrame(count_matrix,index=n_grams,columns=vocabulary)
    return count_matrix

  count_matrix=create_count_matrix(n_pus_1_grams_counts,vocabulary)
  count_matrix+=k
  prob_matrix=count_matrix.div(count_matrix.sum(axis=1),axis=0)
  return prob_matrix,count_matrix
```

```python
def calculate_perplexity(sentence, n_gram_counts, nplus1_gram_counts, n, vocabulary, start_token='<s>', end_token='<e>', k=1.0):
  vocabulary_size = len(vocabulary)
  sentence = [start_token] * n + sentence + [end_token]
  N = len(sentence)
  product_pi = 1.0
  for t in range(n, N):
    n_gram = tuple(sentence[t - n:t])
    word = sentence[t]
    probability = estimate_probability(word, n_gram_counts, nplus1_gram_counts, n_gram, n, vocabulary_size, k=k)
    product_pi *= 1 / probability
  perplexity = product_pi ** (1 / N)
  return perplexity
```

```python
def suggest_word(previous_tokens, n, n_gram_counts, nplus1_gram_counts, vocabulary, k=1.0, start_with=None):
  previous_n_gram = previous_tokens[-n:]
  probabilities = estimate_probabilities(previous_n_gram, n, n_gram_counts, nplus1_gram_counts, vocabulary, k=k)
  suggestion = None
  max_prob = 0
  for word, prob in probabilities.items():
    if start_with is not None and not word.startswith(start_with):
      continue
    if prob > max_prob:
      suggestion = word
      max_prob = prob
  return suggestion, max_prob
```

```python
def get_suggestions(previous_tokens, n_gram_list, vocabulary, k=1.0, start_with=None):
  suggestions = []
  probs = []
  for i in range(len(n_gram_list) - 1):
    n_gram_count = n_gram_list[i]
    n_plus1_gram_count = n_gram_list[i + 1]
    n = i + 1
    suggestion, prob = suggest_word(previous_tokens, n, n_gram_count, n_plus1_gram_count, vocabulary, k=k, start_with=start_with)
    suggestions.append(suggestion)
    probs.append(prob)
  return suggestions, probs
```
